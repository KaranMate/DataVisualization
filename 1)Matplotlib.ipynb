```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Matplotlib Basics
```

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y)
plt.title('Carrier Graph')
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
#Objected Oriented Way
```

```python
#It has a limit of 1 to 4  in third column we have last is limit till 4
plt.subplot(2,2,1)
plt.plot(x**2,x**2)
plt.subplot(2,2,2)
plt.plot(y**2,y**2)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
```

```python
#left, bottom, width, height (the 4 values has sequence )
fig=plt.figure()
axis1=fig.add_axes([0.1,.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.9,0.8,0.4,0.4])
axis2.plot(x,y)
```

```python
#left, bottom, width, height (the 4 values has sequence )
fig=plt.figure(figsize=(10,8),dpi=100) #figsize=(10,8) it is width,height
axis1=fig.add_axes([0.1,.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.6,0.4,0.4])
axis2.plot(x,y)
```
