Boxplot and Violinplot

Boxplots and violin plots display the distribution of data across categories. A boxplot represents quantitative data by showing quartiles within a box and whiskers that extend to depict the rest of the data, excluding outliers identified using the interquartile range.

```python
sns.violinplot(x='day' ,y='total_bill',data=df)
```

Stripplot and Swarmplot

A stripplot shows individual data points for a categorical variable, often complementing box or violin plots to display all observations. A swarmplot works similarly but spaces points to avoid overlap, providing a clearer view of value distribution, though it is less efficient for large datasets.

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day' ,y='total_bill',data=df)
```

```python
#we can combile both swarmplot and violinplot
sns.violinplot(x='tip',y='day',data=df)
sns.swarmplot(x='tip' ,y='day',data=df)
```
