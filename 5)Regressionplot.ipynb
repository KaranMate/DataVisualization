```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
#on x -axis we put the value from which we will predict the data
#from y-axis we put what we want tp predict
#this grapy folloWs the LINEAR ALGORITH Theorem 
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v']
           ,scatter_kws={'s':100})
#markers is use to decide the shape of the graph entity
#scatter_kws is use to decide the sixe of graph entity
```
