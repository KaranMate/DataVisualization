```python
# 1)Which Teams won the most IPL matches?
match_wins=df['match_winner'].value_counts()
sns.barplot(y=match_wins.index,x=match_wins.values,palette="rainbow")
plt.title("IPL 2022 ")
plt.xlabel("No of Match Wins")
plt.ylabel("Teams")
```

```python
#Toss Decision Trend
sns.countplot(x=df['toss_decision'],palette="rainbow")
plt.title("Toss decision Trend")
```

```python
#Toss Winner vs Match Winner
count=df[df['toss_winner'] ==df['match_winner']]['match_id'].count()
percentage=(count*100)/df.shape[0]
percentage.round(2)
```

```python
#How do Team Wins(Runs vs Wicket)
sns.countplot(x=df['won_by'])
```

```python
#Most player of the Match Award
won=df['player_of_the_match'].value_counts().head(10)
won
```

```python
sns.barplot(x=won.values,y=won.index,palette="mako")
plt.title("Top 10 Players of the Match")
plt.ylabel("Player")
plt.xlabel("No of Tropies")
```

```python
#Top Scorer
df.head(2)
```

```python
high=df.groupby("top_scorer")["highscore"].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='barh') #bar is dfault in command barh is for for horizntal
```

```python
#10 Best Bowling figures
df['highest_wickets']=df["best_bowling_figure"].apply(lambda x:x.split('--')[0])
df['highest_wickets']=df['highest_wickets'].astype(int)
top_bowler=df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowler.plot(kind='barh')
```

```python
#Most Matched Played in Venue
venue_count=df['venue'].value_counts()
sns.barplot(x=venue_count.values,y=venue_count.index,palette="rainbow")
```

```python
#Who won the highest margin in Runs
Runs=df[df['won_by'] =="Runs"].sort_values(by ='margin',ascending=False).head(1)[['match_winner','margin']]
Runs
```
